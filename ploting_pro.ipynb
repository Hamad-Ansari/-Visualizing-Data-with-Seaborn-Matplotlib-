## line plot
``A line plot (or line graph) is a fundamental tool in data analysis because it effectively visualizes trends, patterns, and changes over time or ordered categories.``

### When to Use a Line Plot?
✔ Time-series data (e.g., monthly sales)\
✔ Continuous data (e.g., temperature over days)\
✔ Comparing trends (e.g., product A vs. product B performance)

```python
sns.lineplot(x="flipper_length_mm", y="body_mass_g", data=penguins)
```

### When NOT to Use a Line Plot?
✖ Categorical data with no order (use bar charts instead).\
✖ Discrete data points without a trend (scatter plots may be better).

```python
sns.lineplot(x="flipper_length_mm", y="body_mass_g", data=penguins, hue="species")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
line_plot= sns.lineplot(x="flipper_length_mm",
                        y="body_mass_g",
                        data=penguins,
                        hue="species",
                        
                        markers=True,
                        errorbar=None,
                        )
# Add title and labels
plt.title("Flipper Length vs Body Mass")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.legend(title="Species")
# Show the plot
plt.show()
```

### Scatter Plot:
``A scatter plot (or scatter graph) is a powerful tool in data analysis used to visualize the relationship between two numerical variables.``

## 🔍 1. Identify Relationships
Scatter plots help us see if there is a correlation between two variables:

Positive correlation: As one variable increases, the other also increases.

Negative correlation: As one variable increases, the other decreases.

No correlation: The points are scattered randomly.

```python
import numpy as np
```

```python
#scatter plot
plt.scatter(penguins["flipper_length_mm"],
            penguins["body_mass_g"],
            c=penguins["species"].astype("category").cat.codes,
            cmap="viridis",
            
            alpha=0.5,
            edgecolor="k",
            s=100
            )
```

## 📈 2. Detect Patterns or Trends
They can show trends or clusters in the data:

Linear or nonlinear patterns

Clusters of points indicating subgroups

Outliers that don't follow the trend

### Count plot:
``A count plot is a type of bar plot used primarily to show the count of observations in each category of a categorical variable. It's very useful in exploratory data analysis (EDA) when you want to understand the distribution of categorical data.``

## 🧠 When to Use:
You have categorical data (like gender, product type, country).

You want to understand the frequency distribution.

You're preparing data for classification models.



```python
sns.countplot(x="species", data=penguins, palette="pastel")
```

```python
sns.countplot(x="species", data=penguins, palette="pastel")
# Add title and labels
plt.title("Count of Penguins by Species")
plt.xlabel("Species")
plt.ylabel("Count")

# Show the plot
plt.show()
```

## Hist Plot:
``A hist plot (short for histogram plot) is a visualization used to show the distribution of a numerical (continuous) variable by grouping data into bins (intervals) and counting how many values fall into each bin.``

## 🧠 When to Use:
You’re analyzing numerical data.

You want to see how the values are distributed.

You're checking for normality, skewness, or outliers.



```python
# hist plot:
sns.histplot(penguins["flipper_length_mm"],
             bins=20,
             kde=True,
             color="blue",
             alpha=0.5
             )
```

## bar Plot
``A bar plot is a chart used to represent categorical data with rectangular bars, where the length or height of each bar represents the value (like a count, average, or sum) associated with that category``

## 🧠 When to Use:
You want to compare values (not just counts) across categories.

You're working with categorical variables + numerical values.

You want a simple, clear visualization of comparisons.

```python
# bar plot
sns.barplot(x="species",
            y="body_mass_g",
            data=penguins,
            palette="rocket")
```

## 🔁 Bar Plot vs Count Plot:
Count plot is a type of bar plot that automatically counts how many times each category appears.

Bar plot lets you plot any value (like mean, sum, etc.) per category—not just counts.

## Distribution plot
``A distribution plot (or dist plot) is used to visualize the distribution of a single continuous variable — essentially combining a histogram and a kernel density estimate (KDE) curve to show both the frequency and the shape of the data``

## 🧠 When to Use:
You have one numerical variable.

You want to understand how the data is distributed.

You’re looking for patterns, skewness, or outliers.

```python
#distribution plot
sns.displot(penguins["flipper_length_mm"],
            bins=20,
            kde=True,
            color="blue",
            alpha=0.5
            )

```

```python
sns.displot(data=penguins,
            x="flipper_length_mm",
            hue="species",
            element="step")
            
```

```python
sns.displot(data=penguins,
            x="flipper_length_mm",
            hue="species",
            element="step",
            multiple="stack",)
```

| Feature             | Histogram | Distribution Plot (with KDE) |

| Smooth density line | ❌         | ✔️ (with `kde=True`)         |

### density plot
``A density plot is a smoothed version of a histogram. It estimates the probability density function (PDF) of a continuous variable. This means it shows how the data is distributed, but instead of using bars (like a histogram), it uses a smooth curve``

## 🧠 When to Use:
You’re analyzing continuous numerical data.

You want to understand the distribution's shape without the chunkiness of a histogram.

You want to compare multiple distributions visually

```python
sns.displot(data=penguins,
            x="flipper_length_mm",
            hue="species",
            stat='density',
            common_norm=False,
            )
```

### KDE plot

```python
sns.displot(penguins["flipper_length_mm"],
            kind="kde")
```

```python
sns.displot(penguins["flipper_length_mm"],
            kind="kde",
            bw_adjust=0.5,)
```

```python
sns.displot(
    data=penguins,
    x="flipper_length_mm",
    kind="kde",
    hue="species",
    bw_adjust=0.5
)
```

```python
sns.displot(
    data=penguins,
    x="flipper_length_mm",
    kind="kde",
    hue="species",
    multiple="stack",
    fill=True,
)
```

## KDE plot
``A KDE plot (Kernel Density Estimate plot) is a smooth curve that estimates the probability density function (PDF) of a continuous variable. It gives a cleaner, more interpretable view of the data's distribution than a histogram``


```python
sns.kdeplot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
    fill=True,
    common_norm=False,
    alpha=0.5,
    linewidth=0,
)
```

# Heat map
``A heatmap is a data visualization technique that uses color to represent values in a matrix or 2D dataset. It’s commonly used to show correlations, relationships, or frequency distributions between variables.``

## 🧠 When to Use:
You want to understand relationships between multiple variables.

You're analyzing a correlation matrix.

You need to highlight high/low intensities in a matrix.



```python
# heatmap
sns.heatmap(data=penguins.select_dtypes(include='number'))
            
```

### Regression Plot
``A regression plot is a type of data visualization that shows the relationship between two variables and includes a regression line (usually linear) to model that relationship. It’s used to analyze trends, detect correlations, and predict outcomes.``

```python
sns.regplot(x="flipper_length_mm",
            y="body_mass_g",
            data=penguins,
            color="blue",
            marker="o",
            line_kws={"color": "red", "linewidth": 2},
            scatter_kws={"alpha": 0.5, "s": 50}
            )
```

## 🧠 When to Use:
You're analyzing two numerical variables.

You want to model or predict one variable based on another.

You're exploring linear (or simple nonlinear) relationships.
